```python
import pandas as pd
import numpy as np
import json

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('statsfile-2.csv')
df
```

```python
list(df.iloc[:,0].values)
```

```python
pd.json_normalize(list(df.iloc[:,0].values))
```

```python
df1 = pd.json_normalize(
    df['####']
    .apply(json.loads) # the inner JSON is list with the dictionary as the only item
)
```

```python
df1
```

```python
df2 = pd.json_normalize(df1.response.values.tolist())
df2
```

```python
df2.iloc[0,0]
```

```python
df_home= pd.json_normalize(df2.iloc[:,0].values.tolist())
df_home= df_home.add_prefix('home_')
df_home
```

## Calculate home team statistics

```python
df_home_statistics = pd.json_normalize(df_home.iloc[:,0].values.tolist())
df_home_statistics
```

```python
df_new_home_statistics = pd.DataFrame()
for i in df_home_statistics.columns:
    col = pd.json_normalize(df_home_statistics.iloc[:,i].values.tolist())['type'][0]
    df_new_home_statistics[col] =pd.json_normalize(df_home_statistics.iloc[:,i].values.tolist())['value']
df_new_home_statistics = df_new_home_statistics.add_prefix('home_team_')
df_new_home_statistics
```

## Calculate away team statistics

```python
df_away = pd.json_normalize(df2.iloc[:,1].values.tolist())
df_away= df_away.add_prefix('away_')
df_away
```

```python
df_away_statistics = pd.json_normalize(df_away.iloc[:,0].values.tolist())
df_away_statistics
```

```python
df_new_away_statistics = pd.DataFrame()
for i in df_away_statistics.columns:
    col = pd.json_normalize(df_away_statistics.iloc[:,i].values.tolist())['type'][0]
    df_new_away_statistics[col] =pd.json_normalize(df_away_statistics.iloc[:,i].values.tolist())['value']
df_new_away_statistics = df_new_away_statistics.add_prefix('away_team_')
df_new_away_statistics
```

## Add all dataframes to one

```python

df_total = pd.concat([df1 , df_home, df_new_home_statistics , df_away , df_new_away_statistics],axis=1)
df_total
```

```python
df_total.drop(['home_statistics','away_statistics','response','get','errors'],axis=1,inplace=True)
df_total
```

```python
df_total.info()
```

```python
df_total.to_csv('cleaned_stats.csv')
```
